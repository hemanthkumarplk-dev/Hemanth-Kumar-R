# 2.Create histograms and boxplots for numeric features.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset
titanic = pd.read_csv("Titanic-Dataset.csv")

# Select numeric columns
num_features = titanic.select_dtypes(include=['int64', 'float64']).columns

# Create histograms & boxplots
for col in num_features:
    plt.figure(figsize=(12,5))

    # Histogram
    plt.subplot(1,2,1)
    sns.histplot(titanic[col].dropna(), kde=True, bins=30)
    plt.title(f"Histogram of {col}")

    # Boxplot
    plt.subplot(1,2,2)
    sns.boxplot(x=titanic[col].dropna())
    plt.title(f"Boxplot of {col}")

    plt.show()
```

# 3.Use pairplot/correlation matrix for feature relationships.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset
titanic = pd.read_csv("Titanic-Dataset.csv")

# Select numeric columns
num_features = titanic.select_dtypes(include=['int64', 'float64']).columns

# ðŸ“Œ Correlation Matrix + Heatmap
plt.figure(figsize=(10,6))
sns.heatmap(titanic[num_features].corr(), annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Heatmap")
plt.show()

# ðŸ“Œ Pairplot (scatterplots + histograms for all numeric features)
sns.pairplot(titanic[num_features].dropna())
plt.show()
```

 # 4.Identify patterns, trends, or anomalies in the data

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset
titanic = pd.read_csv("Titanic-Dataset.csv")

# 1ï¸âƒ£ Survival rate by Sex
plt.figure(figsize=(6,4))
sns.barplot(x="Sex", y="Survived", data=titanic)
plt.title("Survival Rate by Sex")
plt.show()

# 2ï¸âƒ£ Survival rate by Class
plt.figure(figsize=(6,4))
sns.barplot(x="Pclass", y="Survived", data=titanic)
plt.title("Survival Rate by Class")
plt.show()

# 3ï¸âƒ£ Age distribution by Survival
plt.figure(figsize=(8,5))
sns.histplot(data=titanic, x="Age", hue="Survived", kde=True, bins=30)
plt.title("Age Distribution by Survival")
plt.show()

# 4ï¸âƒ£ Fare distribution for outliers / anomalies
plt.figure(figsize=(8,5))
sns.boxplot(x=titanic["Fare"])
plt.title("Fare Boxplot (to detect outliers)")
plt.show()
```

#  5.Make basic feature-level inferences from visuals.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset
titanic = pd.read_csv("Titanic-Dataset.csv")

# 1ï¸âƒ£ Survival rate by Sex
plt.figure(figsize=(6,4))
sns.barplot(x="Sex", y="Survived", data=titanic)
plt.title("Survival Rate by Sex")
plt.show()

# 2ï¸âƒ£ Survival rate by Pclass
plt.figure(figsize=(6,4))
sns.barplot(x="Pclass", y="Survived", data=titanic)
plt.title("Survival Rate by Pclass")
plt.show()

# 3ï¸âƒ£ Age distribution by Survival
plt.figure(figsize=(8,5))
sns.histplot(data=titanic, x="Age", hue="Survived", kde=True, bins=30)
plt.title("Age Distribution by Survival")
plt.show()

# 4ï¸âƒ£ Fare distribution for outliers
plt.figure(figsize=(8,5))
sns.boxplot(x=titanic["Fare"])
plt.title("Fare Boxplot (to detect outliers)")
plt.show()

# 5ï¸âƒ£ Family members aboard (SibSp + Parch)
plt.figure(figsize=(8,5))
sns.histplot(titanic["SibSp"], bins=10, kde=False)
plt.title("Number of Siblings/Spouses aboard")
plt.show()

plt.figure(figsize=(8,5))
sns.histplot(titanic["Parch"], bins=10, kde=False)
plt.title("Number of Parents/Children aboard")
plt.show()

# -----------------------------
# Feature-level inferences
# -----------------------------
print("\nðŸ”Ž Basic Feature-Level Inferences:")
print("1. Females survived more than males â†’ Sex is a strong survival predictor.")
print("2. 1st class passengers survived more â†’ Pclass influences survival chances.")
print("3. Children and younger passengers had higher survival â†’ Age matters.")
print("4. Fare is skewed with high outliers â†’ wealthier passengers had better survival odds.")
print("5. Most passengers traveled alone (SibSp/Parch=0); large families are rare â†’ potential anomalies.")
```
